```python
def sliding_window_split(text, chunk_size=512, overlap=100):
    chunks = []
    start = 0
    while start < len(text):
        chunks.append(text[start : start + chunk_size])
        start += chunk_size - overlap
    return chunks


```

```python
def retrieve_documents(query, top_k=2, source_filter=None):
    # Filter documents by source before searching
    filtered_documents = [doc for doc in documents if not source_filter or doc.metadata.get("source") == source_filter]
    
    # If no documents left after filtering, return empty list
    if not filtered_documents:
        return []

    # Convert the filtered documents into embeddings
    filtered_embeddings = [embeddings.encode(doc.page_content) for doc in filtered_documents]
    filtered_embeddings = np.array(filtered_embeddings).astype(np.float32)

    # Initialize a FAISS index for the filtered documents
    filtered_index = faiss.IndexFlatL2(filtered_embeddings.shape[1])
    filtered_index.add(filtered_embeddings)

    # Perform the search in the FAISS index for the filtered set of documents
    query_embedding = embeddings.encode(query)
    query_embedding = np.array(query_embedding).astype(np.float32)

    # Perform the search in the filtered FAISS index
    _, indices = filtered_index.search(query_embedding.reshape(1, -1), top_k)

    # Get retrieved documents
    retrieved_docs = [filtered_documents[i] for i in indices[0]]
    return retrieved_docs
```
